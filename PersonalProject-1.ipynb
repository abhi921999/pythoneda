# EDA

```python
df.columns
```

# Gender

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x="Gender",y="Amount", data=sales_gen)
```

# Age

```python
ax=sns.countplot(x='Age Group', data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.barplot( x='Age Group',y='Amount', data= sales_age)
```

# State

```python
state_order=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot( x='State',y='Orders', data= state_order)
```

```python
state_order=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot( x='State',y='Amount', data= state_order)
```

# Marital_Status

```python
ax=sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_order=df.groupby(['Marital_Status',"Gender"],as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot( x='Marital_Status',y='Amount', data= state_order,hue="Gender")
```

# Occupation

```python
ax=sns.countplot(x='Occupation', data=df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_order=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( x='Occupation',y='Amount', data= state_order)
```

# Product Category

```python
ax=sns.countplot(x='Product_Category', data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_order=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( x='Product_Category',y='Amount', data= state_order)
```

```python
state_order=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( x='Product_ID',y='Orders', data= state_order)
```
